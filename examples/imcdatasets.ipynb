```r
download_imc <- function(f, path, force = F) {
    dir.create(path, recursive = T)
    # sce
    sce_path <- file.path(path, "sce.h5ad")
    if (file.exists(sce_path) && force) {
        unlink(sce_path, recursive = T)
    }
    sce <- f(data_type = "sce")
    # TODO: no factor support, convert in colData
    DF <- colData(sce)
    DF[sapply(DF, is.factor)] <- lapply(DF[sapply(DF, is.factor)], as.integer)
    colData(sce) <- DF
    # TODO: no factor support, convert in rowData
    DF <- rowData(sce)
    DF[sapply(DF, is.factor)] <- lapply(DF[sapply(DF, is.factor)], as.character)
    rowData(sce) <- DF
    print(sce)
    adata <- from_SingleCellExperiment(sce, "HDF5AnnData", file=sce_path)
    # masks
    masks_path <- file.path(path, "masks")
    dir.create(masks_path)
    masks <- f(data_type = "masks", on_disk = T, force = force,
    h5FilesPath = masks_path)
    print(masks)
    # images
    images_path <- file.path(path, "images")
    dir.create(images_path)
    images <- f(data_type = "images", on_disk = T, force = force,
    h5FilesPath = images_path)
    print(images)
}
```

```r
download_imc(datafunc, '../data/imcdatasets/Damond_2019_Pancreas', force=T)
```
